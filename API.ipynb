```python
from urllib.request import urlopen
```

```python
api_url =" https://samples.openweathermap.org/data/2.5/forecast?id=524901&appid=b1b15e88fa797225412429c1c50c122a1"
```

```python
url_result = urlopen(api_url)
data = url_result.read()# this comand use for read data
print(data)
```

```python
print(type(data)) # at this point my data in bytes so i convert this data into json 
```

```python
import json
```

```python
json_data = json.loads(data)
```

```python
#print(type(json_data)) # json data show in dict form
print(json_data)
```

```python
json_data['list']
```

```python
json_string = json.dumps(json_data) # dumps use for convert dict to string 
print(json_string)
```

# Facebook API

```python
import requests
url = "http://graph.facebook.com/4/picture?type=large"
```

```python
r = requests.get(url)
print(r.content)# r.content give me all url content in bineray form
```

```python
with open("sample_pic.jpg","wb") as f:
    f.write(r.content) # this convert bineray to jpg
```
